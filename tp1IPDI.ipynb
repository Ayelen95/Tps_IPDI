```python
##### TP1 IPDI
import tkinter as tk
from tkinter import ttk, Frame

import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

import numpy as np 

import imageio
import imageio.v2 as imageio

imagenTrabajad = "lena128C.png"

def escalar(matriz,a=0.0, b=0.0):
    #Yprima = matriz[:,:,0] * float(a)
    #Iprima = matriz[:,:,1] * float(b)
    #Qprima = matriz[:,:,2] * float(b)
    #print(valor_a_entry.get())
    #auxa = float(valor_a_entry.get())
    #auxb = float(valor_b_entry.get())
    h,w,c=matriz.shape # obtenemos el tamaño de la imagen original
    mrYIQ = np.zeros((h,w,3)) # creamos una matriz donde generar la imagen
    mrYIQ[:,:,0] = matriz[:,:,0] * a
    mrYIQ[:,:,1] = matriz[:,:,1] * b
    mrYIQ[:,:,2] = matriz[:,:,2] * b
    return mrYIQ
    

def Calcular():
    Paso1 = RGB_to_YIQ(imagenTrabajad)
    #print(Paso1) #Hecho
    matriz = escalar(Paso1)
    #print(matriz)
    Paso2 = np.clip(matriz/1.,0,1)
    plt.title("Lena YIQ")
    plt.imshow(Paso2)
    mostrarDibujo(Paso2,1,5)
    matrizLast = YIQ_to_RGB(Paso2)
    print(matrizLast)
    Paso7 = np.clip(matriz/1.,0,255)
    plt.title("Lena RGB")
    plt.imshow(Paso7)
    mostrarDibujo(Paso7,1,6)
    #plt.imshow(np.clip((normalize(matrizLast))/1,0,255))
    #print(np.clip((normalize(matrizLast))/1,0,255))    
    
##Funcion  de RGB A YIQ
def RGB_to_YIQ(imgSinNorm):
    im = imageio.imread(imgSinNorm)/255.0
    h,w,c=im.shape # obtenemos el tamaño de la imagen original
    mYIQ = np.zeros((h,w,3)) # creamos una matriz donde generar la imagen
    mYIQ[:,:,0] = 0.299*im[:,:,0] + 0.587*im[:,:,1] + 0.114*im[:,:,2]
    mYIQ[:,:,1] = 0.595716*im[:,:,0] + -0.274453*im[:,:,1] + -0.321263*im[:,:,2]
    mYIQ[:,:,2] = 0.211456*im[:,:,0] + -0.522591*im[:,:,1] + 0.311135*im[:,:,2]
    return mYIQ 

def escalar(matriz):
    #Yprima = matriz[:,:,0] * float(a)
    #Iprima = matriz[:,:,1] * float(b)
    #Qprima = matriz[:,:,2] * float(b)
    #print(valor_a_entry.get())
    h,w,c=matriz.shape # obtenemos el tamaño de la imagen original
    mrYIQ = np.zeros((h,w,3)) # creamos una matriz donde generar la imagen
    mrYIQ[:,:,0] = matriz[:,:,0] *(a.get())
    mrYIQ[:,:,1] = matriz[:,:,1] *(b.get())
    mrYIQ[:,:,2] = matriz[:,:,2] *(b.get())
    return mrYIQ

##Funcion  de YIQ A RGB
def YIQ_to_RGB(valoresMatriz):
    im = valoresMatriz
    h,w,c=im.shape # obtenemos el tamaño de la imagen original
    mRGB = np.zeros((h,w,3)) # creamos una matriz donde generar la imagen
    mRGB[:,:,0] = 1*im[:,:,0] + 0.9663*im[:,:,1] + 0.6210*im[:,:,2]
    mRGB[:,:,1] = 1*im[:,:,0] + -0.2721*im[:,:,1] + -0.6474*im[:,:,2]
    mRGB[:,:,2] = 1*im[:,:,0] + -1.1070*im[:,:,1] + 1.7046*im[:,:,2]
    return mRGB

def normalize(img):
    lmin = float(img.min())
    lmax = float(img.max())
    return np.floor((img-lmin)/(lmax-lmin)*255.0)

def mostrarDibujo(valorM,c,r):
    frame = Frame(root, bg='gray22',bd=3)
    frame.grid(column=c,row=r)
    imshow = plt.imshow(valorM)
    figure = plt.figure()
    chart = FigureCanvasTkAgg(figure, master=frame)
    chart.draw()
    chart.get_tk_widget().grid(row = r, column=c, pady = 5)
    
    #img = tk.PhotoImage(valorM, master=root)
    #img = img.subsample(8, 8)
    #imagen_aux = ttk.Label(root, image=img)
    #imagen_aux.grid(column=c, row=r )
    #plt.grid()
    #axes = plt.axes()
    #axes.set_xlim([0, 130])
    #axes.set_ylim([130, 30])
    return None
# root window
root = tk.Tk()
root.geometry("700x800")
root.title('Procesamiento Digital de Imágenes - TP1 - 2022')
root.resizable(0, 0)

# configure the grid
root.columnconfigure(0, weight=1)
root.columnconfigure(1, weight=2)

# Valores dados por el usuario
a = tk.DoubleVar(value=1.0)
b = tk.DoubleVar(value=0.0)

# label valor a
valor_a_label = ttk.Label(root, text="Valor a:")
valor_a_label.grid(column=0, row=0, sticky=(tk.E), padx=5, pady=10)

valor_a_entry = ttk.Entry(root, textvariable=a, width=40)
valor_a_entry.grid(column=1, row=0, sticky=tk.N, padx=5, pady=10)

# label valor b
valor_b_label = ttk.Label(root, text="Valor b:")
valor_b_label.grid(column=0, row=1, sticky=tk.E, padx=5, pady=10)

valor_b_entry = ttk.Entry(root,textvariable=b, width=40)
valor_b_entry.grid(column=1, row=1, sticky=tk.N, padx=5, pady=10)

# salir button
salir_button = ttk.Button(root, text="Salir", command = root.destroy)
salir_button.grid(column=0, row=3, sticky=tk.N, padx=5, pady=5)

# calcular button al fondo
calcular_button = ttk.Button(root, text="Calcular", width=20, command=Calcular)
calcular_button.grid(column=1, row=3, sticky=tk.N, padx=5, pady=5)
#calcular_button.bind("<Return>", Calcular)

# label valor aux
valor_aux_label = ttk.Label(root, text="")
valor_aux_label.grid(column=0, row=4, sticky=(tk.E), padx=5, pady=10)

# subo imagen de lena.png
img = tk.PhotoImage(file=imagenTrabajad, master=root)
#img = img.subsample(8, 8)
imagen_label = ttk.Label(root, image=img)
imagen_label.grid(column=0, row=5)



root.mainloop()
```
